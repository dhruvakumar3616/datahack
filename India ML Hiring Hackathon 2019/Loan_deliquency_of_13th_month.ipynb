### Define a  function for converting date to required format for object datatype

```python
import datetime

# train data
#origination_date
def extract_origination_date(date):
    return datetime.datetime.strptime(date, '%Y-%m-%d')

#first_payment
def extract_first_payment(date):
    return datetime.datetime.strptime(date, '%m/%Y')
```

## convert attributes to required format

```python
# list the corresponding attributes as per datatype 

#category
cat_cols = ['loan_id', 'source', 'financial_institution','loan_term', 'number_of_borrowers', 
 'loan_purpose', 'insurance_percent',  'insurance_type',
 'm1','m2', 'm3', 'm4', 'm5', 'm6', 'm7', 'm8', 'm9', 'm10','m11', 'm12']

#float
float_cols=['interest_rate', 'unpaid_principal_bal']

#integer
int_cols = ['borrower_credit_score','debt_to_income_ratio', 'loan_to_value','co-borrower_credit_score']


#date
date_cols =[ 'origination_date', 'first_payment_date',]
```

```python
# convert attributes to required data type of train and test data

#categorical
train_df[cat_cols] = train_df[cat_cols].astype('category')
test_df[cat_cols] = test_df[cat_cols].astype('category')

#iteger
train_df[int_cols] = train_df[int_cols].astype('int64')
test_df[int_cols] = test_df[int_cols].astype('int64')

#float
train_df[float_cols] = train_df[float_cols].astype('float64')
test_df[float_cols] = test_df[float_cols].astype('float64')

# target attribute
train_df['m13'] = train_df['m13'].astype('category')
```

```python
# date attributes
#origination_date
train_df.origination_date = train_df.origination_date.apply(extract_origination_date)

# first_payment_date
train_df.first_payment_date = train_df.first_payment_date.apply(extract_first_payment)
```

## Create training and validating data 
As it is imbalanced data set, we hae to use __stratified sampling based on "m13"__ attribute for train and validation set.

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=333) 
```

```python
for train_index, valid_index in split.split( train, train["m13"]): 
    strat_train_set = train.iloc[train_index] 
    strat_valid_set = train.iloc[valid_index]
```

# Model building

```python
# function to display scores
from sklearn.metrics import f1_score,accuracy_score,confusion_matrix, recall_score, precision_score

# function to display scores
def display_scores(y_true,y_pred):
    print("Accuracy = ", accuracy_score(y_true,y_pred))
    print("Recall = ", recall_score(y_true,y_pred))
    print("Precision = ",precision_score(y_true,y_pred))
    print("F1 Score = ",f1_score(y_true,y_pred))
```

## Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier()
```

### Stratified sample sets

```python
# target
st_train_y = strat_train_set.m13

# predixtors
st_train = strat_train_set.drop('m13',1)
```

```python
#target
st_valid_y = strat_valid_set.m13

#predictors
st_valid = strat_valid_set.drop('m13',1)
```

# train model on stratified training data

```python
dt_clf.fit(st_train,st_train_y)
```

```python
# predict on training data
display_scores(st_train_y,dt_clf.predict(st_train))
```

```python
# predict on validation data
display_scores(st_valid_y,dt_clf.predict(st_valid))
```

```python
# predict on test data
predict_test_dt_1 = dt_clf.predict(test)
```

```python
# convert prediction on test data to data frame
predict_test_dt_1 = pd.DataFrame(predict_test_dt_1)
predict_test_dt_1.index = test_df.loan_id
predict_test_dt_1.columns = ['m13']
predict_test_dt_1.head()
```
