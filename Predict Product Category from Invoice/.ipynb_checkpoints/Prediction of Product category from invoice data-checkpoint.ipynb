# Predict Product Category from Invoice


Organizations around the world purchase goods & services from their suppliers via Purchase Order-Invoice exchange. Organizations raise Purchase Order for specific item and expected price. Suppliers then raise an Invoice to the organization for billing. As part of process streamlining, organizations catalogue items, i.e., they document all the item details in their procurement system. However in many cases organizations are required to purchase non-catalogued items.

***

# Task:
 Need to innovate a  **High Accuracy Classification system** to predict **‘Product Category’** from the given invoice information.

***

# Data Description

### Data consists of Invoice details for multiple customers as described below:


|Column Name|Expansion|Description                      |
|:-|-|-|
|Inv_ID | Invoice ID|Unique number representing Invoice created by supplier/vendor |
|Vendor Code |Vendor ID | Unique number representing Vendor/Seller in the procurement system |
|GL_Code| Account’s Reference ID | - |
|Inv_Amt |Invoice Amount | - |
|Item Description|Description of Item Purchased | - |
|Product Category|Category of Product for which Invoice is raised|A pseudo product category is represented in the dataset as CLASS-???, where ? is a digit.|

# 4. Prepare data for model building

## 4.1 Define a function to preprocess the text in the train and test data

```python
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\D+')
stop_words =stopwords.words('english')

def pre_process_text(text):
    text = tokenizer.tokenize(text)                  #tokenize words
    text = [x.lower()  for x in text ]               #convert all words into lower case
    text = [x for x in text if x not in stop_words]  # remove all stopwords
    return " ".join(text)                            #join all words with a space between words
```

## 4.2 Apply the 'pre_process_text()' function to all instances in train and test data

```python
train = train_df.Item_Description.apply(pre_process_text)
test = test_df.Item_Description.apply(pre_process_text)
```

## 4.3 Create TF-IDF matrix for train data for training the model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

### 4.3.1 Create vectorizer

```python
vectorizer = TfidfVectorizer()
#vectorizer = TfidfVectorizer(ngram_range=(3, 3), min_df=0.001)
```

```python
# create tfidf matrix from train data 
# call vectorizer.fit_transform() to train i.e create features from training data
#train data
train_X = vectorizer.fit_transform(train)
```

```python
# use the features generated from train data and create tf-idf matrix for test data
# call vectorizer.transform() to use the features generated by training data
#test data
test_X = vectorizer.transform(test)
```

```python
type(train_X)
```

### 4.3.2 convert sparse tf-idf matrix to Dense matrix

```python
train_tf_idf_matrix = train_X.todense()
test_tf_idf_matrix = test_X.todense()
```

### 4.3.3 Shape of the tf-idf matrix

```python
train_tf_idf_matrix.shape
```

```python
test_tf_idf_matrix.shape
```

### 4.3.4 Features created i.e. words in the train data

```python
vectorizer.get_feature_names()
```

### 4.3.5 Lets see the TF-IDF matrix with features names

```python
train_tf_idf_df = pd.DataFrame(train_tf_idf_matrix)
train_tf_idf_df.columns = vectorizer.get_feature_names()
```

```python
test_tf_idf_df = pd.DataFrame(test_tf_idf_matrix)
test_tf_idf_df.columns = vectorizer.get_feature_names()
```

```python
train_tf_idf_df.head(10)
```

```python
test_tf_idf_df.head(10)
```

# Target column

```python
y = train_df.Product_Category
```

# 5. Split the data into train and validation set from TF-IDF matrix of train data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_tf_idf_df,y,test_size=0.3, random_state=42)
```

### Shape of training and validation set 

```python
print("X_train = {}\n X_test={}\n y_train={}\n y_test={}"
      .format(X_train.shape, X_test.shape, y_train.shape, y_test.shape))
```

```python
from sklearn.naive_bayes import GaussianNB
NB_clf = GaussianNB()
```

```python

NB_clf.fit(X_train,y_train)
```

### 6.1.2 Precict on validation set

```python
pred = NB_clf.predict(X_test) 
```

### 6.1.3 Accuracy on validation set

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

### 6.1.4 Confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm =confusion_matrix(y_test,pred)
print(cm)
```
